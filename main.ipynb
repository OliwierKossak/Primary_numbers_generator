```python
import json
import os
import matplotlib.pyplot as plt
from basic_algorithm import PrimeNumbersGenerator
from my_algorithm import PrimeNumberAlogrithm

range_number = 20000
prime_numbers_generator = PrimeNumbersGenerator(range_number)
prime_numbers_generator2 = PrimeNumberAlogrithm(range_number)
time1 = round(prime_numbers_generator.measure_time(), 3)
time2 = round(prime_numbers_generator2.measure_time(), 3)
algorithm_improved = round(abs(time2 / time1 * 100 - 100), 2)

print(f"The algorithm execution time has been reduced by: {algorithm_improved}%")


```

```python
plt.bar(['basic algorithm', 'my algorithm'], [time1, time2])
plt.xlabel("algorithms")
plt.ylabel('execution time in seconds')
plt.show()

```
